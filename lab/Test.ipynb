```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import numpy as np
import scipy.stats
from matplotlib import pyplot
pyplot.rcParams['figure.figsize'] = [15, 5]
```

```python
X = np.random.rand(10000)
y = X*2-1
```

```python
pyplot.scatter(X, y)
pyplot.show()
```

```python
pearson, _ = scipy.stats.pearsonr(X, y)
print(pearson)
```
