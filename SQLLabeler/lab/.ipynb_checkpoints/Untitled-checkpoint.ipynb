```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import numpy as np
import scipy.stats
from matplotlib import pyplot
```

```python
X = np.random.rand(100)
y = np.sin(X*np.pi*20)
```

```python
pyplot.scatter(X, y)
pyplot.show()
```

```python
pearson, _ = scipy.stats.pearsonr(X, y)
print(pearson)
```
